# STEP #1: PROBLEM STATEMENT


- 유방암진단 문제 : 양성이냐 악성이냐 예측   
- 30 features are used, examples:
        - radius (반지름)
        - texture (조직)
        - perimeter (둘레)
        - area
        - smoothness (local variation in radius lengths)
        - compactness (perimeter^2 / area - 1.0)
        - concavity (오목함)
        - concave points (오목한 부분의 점)
        - symmetry  (대칭)
        - fractal dimension ("coastline approximation" - 1)

- 30 input features
- Number of Instances: 569
- Class Distribution: 212 Malignant(악성), 357 Benign(양성)
- Target class:
         - Malignant(악성)
         - Benign(양성)


https://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Wisconsin+(Diagnostic)

# STEP #4: MODEL TRAINING (FINDING A PROBLEM SOLUTION)

```python
df.isna().sum()
```

```python
df.head(3)
```

```python
y = df['target']
```

```python
X = df.iloc[ : , 0 : -2+1 ]
```

```python
X.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler_X = StandardScaler()
```

```python
X = scaler_X.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
from sklearn.svm import SVC
```

```python
classifier1 = SVC(kernel='linear', random_state=2)
```

```python
classifier2 = SVC(kernel='rbf', random_state=2)
```

```python
classifier1.fit(X_train, y_train)
```

```python
classifier2.fit(X_train, y_train)
```

# STEP #5: EVALUATING THE MODEL

```python
y_pred1 = classifier1.predict(X_test)
```

```python
y_pred2 = classifier2.predict(X_test)
```

```python
y_pred1
```

```python
y_pred2
```

```python
y_test.values
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
confusion_matrix(y_test, y_pred1)
```

```python
accuracy_score(y_test, y_pred1)
```

```python
confusion_matrix(y_test, y_pred2)
```

```python
accuracy_score(y_test, y_pred2)
```

# STEP #6: IMPROVING THE MODEL

```python
# Grid Search 
```

```python
param_grid = { 'kernel' : ['linear','rbf','poly'] , 'C' : [0.1, 1, 10] ,
             'gamma' :  [0.01, 0.1, 1] }
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV( SVC() , param_grid , refit = True, verbose= 4)
```

```python
grid.fit(X_train, y_train)
```

```python
classifier = grid.best_estimator_
```

```python
classifier
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = classifier.predict(X_test)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```
